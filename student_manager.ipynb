```python
import csv
import os

FILE = "data.csv"

def init_file():
    if not os.path.exists(FILE):
        with open(FILE, 'w', newline='') as f:
            writer = csv.writer(f)
            writer.writerow(["ID", "Name", "Age", "Course"])

def add_student():
    init_file()
    sid = input("Enter ID: ")
    name = input("Enter Name: ")
    age = input("Enter Age: ")
    course = input("Enter Course: ")

    with open(FILE, 'a', newline='') as f:
        writer = csv.writer(f)
        writer.writerow([sid, name, age, course])

    print("Student added successfully!")

def view_students():
    init_file()
    with open(FILE, 'r') as f:
        reader = csv.reader(f)
        for row in reader:
            print(row)

def search_student():
    sid = input("Enter ID to search: ")
    found = False

    with open(FILE, 'r') as f:
        reader = csv.reader(f)
        for row in reader:
            if row[0] == sid:
                print("Found:", row)
                found = True

    if not found:
        print("Student not found")

def update_student():
    sid = input("Enter ID to update: ")
    rows = []
    found = False

    with open(FILE, 'r') as f:
        reader = csv.reader(f)
        for row in reader:
            if row[0] == sid:
                print("Enter new details:")
                name = input("Name: ")
                age = input("Age: ")
                course = input("Course: ")
                rows.append([sid, name, age, course])
                found = True
            else:
                rows.append(row)

    with open(FILE, 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerows(rows)

    if found:
        print("Updated successfully!")
    else:
        print("Student not found")

def delete_student():
    sid = input("Enter ID to delete: ")
    rows = []
    found = False

    with open(FILE, 'r') as f:
        reader = csv.reader(f)
        for row in reader:
            if row[0] != sid:
                rows.append(row)
            else:
                found = True

    with open(FILE, 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerows(rows)

    if found:
        print("Deleted successfully!")
    else:
        print("Student not found")
```
